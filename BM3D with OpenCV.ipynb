```python
import os
import cv2
import bm3d
import numpy as np
from skimage import io, img_as_float
from matplotlib import pyplot as plt
from sklearn.metrics import mean_squared_error
from skimage.metrics import peak_signal_noise_ratio
import glob
```

```python
path="test/CNV/*"
```

```python
data_dir_path = u"test/drusen1/"
file_list = os.listdir(r'test/drusen1/')
data_dir_path2 = u"filtered/fDrusen/"
a=0
for file_name in file_list:
    root, ext = os.path.splitext(file_name)
#         if ext == u'.png' or u'.jpeg' or u'.jpg':
    abs_name = data_dir_path +'/' + file_name
    img=img_as_float(io.imread(abs_name,as_gray=True))
    denoised= bm3d.bm3d(img, sigma_psd=30/255, stage_arg=bm3d.BM3DStages.HARD_THRESHOLDING)
    abs_name2 = data_dir_path2 +'/' + file_name
    # cv2.imwrite(str(a)+'., denoised)
    io.imsave(abs_name2,denoised)
    print(f"filtered by BM3D No:{a}")
    a+=1
```
