```python
import os
import cv2
import bm3d
import numpy as np
```

```python
def main():
    data_dir_path = u"test/CNV/"
    file_list = os.listdir(r'test/CNV/')
    data_dir_path2 = u"filtered/CNV"
    for file_name in file_list:
        root, ext = os.path.splitext(file_name)
        if ext == u'.png' or u'.jpeg' or u'.jpg':
            abs_name = data_dir_path +'/' + file_name
            img = cv2.imread(abs_name)
            denoised= bm3d.bm3d(img, sigma_psd=30/255, stage_arg=bm3d.BM3DStages.HARD_THRESHOLDING)
            abs_name2 = data_dir_path2 +'/' + file_name
            cv2.imwrite(abs_name2, denoised)
if __name__ =='__main__':
    main()
```
