```python
import cv2
import numpy as np
from skimage import io,img_as_float
from skimage.restoration import denoise_nl_means, estimate_sigma

from matplotlib import pyplot as plt
```

```python
img =cv2.imread("test/CNV-163081-6.jpeg")
```

```python
sigma_est=np.mean(estimate_sigma(img, multichannel=True))
```

```python
denoise_img = denoise_nl_means(img, h=1. *sigma_est, fast_mode=False,
                               patch_size=5,
                               patch_distance=3,
                               multichannel=False)
```

```python

cv2.imshow("original", img)
cv2.imshow("denoised with NLM", denoise_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
