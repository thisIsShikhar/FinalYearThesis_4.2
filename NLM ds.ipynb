```python
import numpy as np
from matplotlib import pyplot as plt
from skimage import img_as_ubyte, img_as_float
from matplotlib import pyplot as plt
from skimage import io
import numpy as np
import cv2
from skimage.metrics import peak_signal_noise_ratio

from sklearn.metrics import mean_squared_error
# fastNlMeansDenoising(InputArray src, OutputArray dst, float h=3, int templateWindowSize=7, int searchWindowSize=21 )
```

```python
noisy_img = img_as_ubyte(io.imread("test/CNV-163081-6.jpeg", as_gray=False))  #Only 8 bit supported for CV2 NLM
# ref_img = io.imread("images/MRI_images/MRI_clean.tif")
```

```python
img= img_as_ubyte(noisy_img)
```

```python
NLM_CV2_denoise_img = cv2.fastNlMeansDenoising(img, None, 34, 38, 25)
```

```python
# plt.imsave("images/MRI_images/NLM_CV2_denoised.tif", NLM_CV2_denoise_img, cmap='gray')
plt.imshow(NLM_CV2_denoise_img, cmap='gray')
```

```python
anisotropic_cleaned_psnr = peak_signal_noise_ratio(noisy_img, NLM_CV2_denoise_img)
```

```python
print(anisotropic_cleaned_psnr )
```

```python
a=mean_squared_error(noisy_img, NLM_CV2_denoise_img)
```
